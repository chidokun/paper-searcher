```python
from bs4 import BeautifulSoup
import requests
```

```python
link = 'https://paperswithcode.com/search?q_meta=&q_type=&q={}'
```

```python
response = requests.get(link.format("MetaFormer Is Actually What You Need for Vision"))
response.text
```

```python
parser = BeautifulSoup(response.text, 'html.parser')
```

```python
container = parser.find('div', { 'class': 'infinite-container'})
items = container.findAll(class_='item')
len(items)
```

```python
def extract(item):
    title = item.find('h1')
    title = title.find('a')
    print(title)
    return {
        "title": title.text,
        "link": "https://paperswithcode.com" + title["href"]
    }
```

```python
[ extract(item) for item in items ] 
```
